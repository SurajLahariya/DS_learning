```python
import pandas as pd
import numpy as np
```

```python
url = 'https://raw.githubusercontent.com/SurajLahariya/DS_learning/main/Book.csv'
```

```python
df=pd.read_csv(url, encoding= 'unicode_escape')
```

```python
df
```

```python
book=df.iloc[:,1:]
```

```python
book.head()
```

```python
book.columns=['User_id', 'Book_title', 'Book_rating']
book.head(2)
```

```python
book.info()
```

```python
book.sort_values(by=['User_id'])
```

```python
len(book.User_id.unique())
```

```python
len(book.Book_title.unique())
```

```python
book[book.duplicated()]
```

```python
book.drop_duplicates(subset=None, keep='first', inplace=True)
```

```python
book[book.duplicated()]
```

```python
Book=book.drop_duplicates(subset=["Book_title"])
```

```python
Book.info()
```

```python
user_book=Book.pivot(index='User_id', columns='Book_title', values='Book_rating').reset_index(drop=True)
```

```python
user_book
```

```python
user_book.fillna(0, inplace=True)
```

```python
user_book.info()
```

```python
user_book.index=Book.User_id.unique()
```

```python
user_book
```

```python
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim = 1 - pairwise_distances(user_book.values,metric='cosine')
```

```python
user_sim
```

```python
user_sim_df = pd.DataFrame(user_sim)
user_sim_df
```

```python
user_sim_df.index = Book.User_id.unique()
user_sim_df.columns = Book.User_id.unique()
```

```python
user_sim_df.iloc[0:5, 0:5]
```

```python
user_sim_df.idxmax(axis=1)[0:5]
```
